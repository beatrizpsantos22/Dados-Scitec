```python

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.histplot(df['gasolina_comum_preco_revenda_avg'], bins=11, kde=True, ax=axes[0])
axes[0].set_title('Distribuição do preço médio de revenda da gasolina comum')
axes[0].set_xlabel('Preço Médio')
axes[0].set_ylabel('Frequência')


sns.histplot(df['oleo_diesel_preco_revenda_avg'], bins=30, kde=True, ax=axes[1], color='skyblue')
axes[1].set_title('Distribuição do preço médio de revenda do óleo diesel')
axes[1].set_xlabel('Preço Médio')
axes[1].set_ylabel('Frequência')

# Ajusta o layout para não haver sobreposição
plt.tight_layout()
plt.show()
```

```python
numerical_cols = ['ano', 'gasolina_comum_preco_revenda_avg', 'oleo_diesel_preco_revenda_avg']

# Cria o box plot usando seaborn
sns.boxplot(data=df[numerical_cols])
plt.title('Distribuição e Escala das Variáveis Numéricas')
plt.ylabel('Valor')
plt.xlabel('Variáveis')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()

# Aplicando StandardScaler
df_standard_scaled = df[numerical_cols].copy()
df_standard_scaled[numerical_cols] = standard_scaler.fit_transform(df[numerical_cols])

# Aplicando MinMaxScaler
df_minmax_scaled = df[numerical_cols].copy()
df_minmax_scaled[numerical_cols] = minmax_scaler.fit_transform(df[numerical_cols])
```

```python
sns.boxplot(data=df_standard_scaled[numerical_cols])
plt.title('Distribuição e Escala das Variáveis Numéricas')
plt.ylabel('Valor')
plt.xlabel('Variáveis')
plt.show()
```

```python
sns.boxplot(data=df_minmax_scaled[numerical_cols])
plt.title('Distribuição e Escala das Variáveis Numéricas')
plt.ylabel('Valor')
plt.xlabel('Variáveis')
plt.show()
```

```python
sns.boxplot(x='ano', y='gasolina_comum_preco_revenda_avg', data=df, palette='pastel')
plt.title('Preço Médio por Ano')
plt.xlabel('Ano')
plt.ylabel('Preço Médio')
plt.show()
```

```python
sns.scatterplot(x='mes', y='gasolina_comum_preco_revenda_avg', data=df, alpha=0.6)
plt.title('Relação entre Preço Médio e Mês')
plt.xlabel('Mês')
plt.ylabel('Média de Preço')
plt.show()
```

```python
sns.countplot(x='Urban_or_Rural', hue='Primary_Device', data=df, palette='magma')
plt.title('Uso de Dispositivo por Localização (Urbano/Rural)')
plt.xlabel('Localização')
plt.ylabel('Contagem')
plt.legend(title='Dispositivo')
plt.show()
```
